### Method

Loading Forest cover type data-set & importing dependencies.

```python
import numpy as np
import abc


##your method implementation goes here

# Importing Dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm, metrics
from sklearn.preprocessing import StandardScaler, LabelBinarizer
from sklearn.pipeline import make_pipeline
from sklearn.svm import LinearSVC, SVC
from sklearn.metrics import plot_confusion_matrix, roc_auc_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


# Loading the dataset
data=pd.read_csv("./covtype.csv", sep=',')

# Original number of samples
data.shape
```

Due to speed issues, i decided to decrease the number of samples to '100000'.<br>
Although you can increase the number of samples if your machine is fast.

```python
# Due to speed issues, i decided to decrease the number of samples to '100000'.
# Although you can increase the number of samples if your machine is fast.
data = data[:100000]
data.shape
```

Loading & splitting data for the 'Cover_Type' as output

```python
# Loading & splitting data for the 'Cover_Type' as output
x_train=data.drop(labels=['Cover_Type'],axis=1)
y_train=data['Cover_Type'].astype(int)
X_train,X_test,Y_train,Y_test = train_test_split(x_train,y_train,test_size=0.2, random_state=2)
X_train = np.array(X_train)
X_test = np.array(X_test)
Y_train = np.array(Y_train)
Y_test = np.array(Y_test)
```

Defining a utility function to help & calculate ROC score for Multi-label output (using 'LabelBinarizer()').

```python
# Defining a utility function to help & calculate ROC(AUC) score for Multi-label output (using 'LabelBinarizer()')
def multi_label_roc(y_test, y_pred, average="macro"):
    LabelBinarizer_func = LabelBinarizer()
    LabelBinarizer_func.fit(y_test)
    y_test = LabelBinarizer_func.transform(y_test)
    y_pred = LabelBinarizer_func.transform(y_pred)
    return roc_auc_score(y_test, y_pred, average=average)
```

___

**Support Vector Machine**: Although quite old, SVM still happens to be one of the best plain-ML (i.e. Non-Neural network) algorithm. It's idea is to transform non linearly separable data into a different dimension, where SVM is able to find a hyper-plane that can linearly seperate the data. It uses the concept of equidistant vectors supporing this hyper-plane. These vectors are called supporting vectors. This is a simple explanation. More complex versions exist with soft-margin SVM, multi-class SVM, etc.

___
For **Metrics** i chose the following:

1) **Accuracy Score**: As the obvious name suggests, it measures the overall accuracy of the trained model. It's one of the most important metrics for analyzing 'Classification models'.

2) **ROC (AUC) Score**: ROC stands for 'Receiver Operating Characteristic' & AUC for 'Area under the curve'. It's another important metrics used for analyzing 'Classification models'. However, it's originally suited for binary outputs. Thus, i had to use 'LabelBinarizer()' of sklearn to convert it into binary labels for each label. ROC(AUC) measures the area under the curve for 'precision vs recall' graph. Anything b/w 0.5 to 1 is considered as a good result, with more the merrier. It tries to compare the accuracy of model with respect to marginal of True Positive, False Positive, True Negative, False Negative.

3) **Confusion Matrix**: It's another important metric for analyzing 'Classification models'. It follows a similar idea of ROC. Hence it tries to compare the accuracy & False belief of model for different classes individually. In confusion matrix, one axis represent the Ground truth label-class & the other axis represents the predicted label-class. Thus, the matrix gives you an idea about the overall accuracy of model for each class. We also plot the confusion matrix using visualization tools.

___
Firstly, I'll implementing SVM for the original data-size.
I train a new SVM (Linear & Kernel) -> Test model on testing data-set -> plot the metrics ('Accuracy score','ROC (AUC) score', 'Confusion Matrix')

```python
# Initialize SVM (Linear) & pipeline (Standardizing mean & variance)
clf = make_pipeline(StandardScaler(),LinearSVC(random_state=0, tol=1e-4))

# Train SVM
clf.fit(X_train, Y_train)

# Predict on Testing set
y_pred = clf.predict(X_test)

# Plot Confusion Matrix using trained classifier on Test Set.
plot_confusion_matrix(clf, X_test, Y_test)

# Printing the data-size (i.e. i)
print('Printing Accuracy & ROC for original data-size')

# Calculate & print 'Accuracy' score using the predicted values
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))

# Calculate & print 'ROC' score using the predicted values
roc = multi_label_roc(Y_test, y_pred)
print("ROC:",roc)
```

```python
# Initialize SVM (Kernel) & pipeline (Standardizing mean & variance)
clf = make_pipeline(StandardScaler(),SVC(kernel="poly", degree=2, C=1))

# Train SVM
clf.fit(X_train, Y_train)

# Predict on Testing set
y_pred = clf.predict(X_test)

# Plot Confusion Matrix using trained classifier on Test Set.
plot_confusion_matrix(clf, X_test, Y_test)

# Printing the data-size (i.e. i)
print('Printing Accuracy & ROC for original data-size')

# Calculate & print 'Accuracy' score using the predicted values
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))

# Calculate & print 'ROC' score using the predicted values
roc = multi_label_roc(Y_test, y_pred)
print("ROC:",roc)
```

___
**Principal component analysis**: PCA tries to find the Principal components (i.e. directions) that have the maximum variance for the projection of data-points. Thus, these Principal components carry the most information embedded in the data-set.

We can use this idea to reduce the dimensionality of data-set such that a lower dimensional data looses the least amount of information after reducing dimensions. Thus, data-size is reduced heavily in comparison to actual information (which is lost as minimum as possible.). Thses Principal components are orthogonal to each other, thus they are not correlated & provide unique contribution to the data.
___

Then i'll perform PCA Dimensionality reduction for different data-sizes (i.e. 10%,15%,20%,25%, & 30% of the originial data-size).

Later i'll get a new trainting & testing data-set for the dimensionally reduced data-set.
I train a new SVM -> Test model on testing data-set -> plot the metrics ('Accuracy score','ROC (AUC) score', 'Confusion Matrix')

```python
# Approximate number of components (out of total 55) in order to get 10%,15%,20%,25%, & 30% of the originial data-size
n_comps = [5,7,10,13,15]

# Loop to iterate through different dimensionality reduction
for ind,i in enumerate(n_comps):
    
    # Initialize PCA with number of components ='5*i' (i.e. Approximately 10%,15%,20%,25%, & 30% of the originial 55 components)
    model = PCA(n_components=i)
    model.fit(X_train)
    X_train_new = model.transform(X_train)
    X_test_new = model.transform(X_test)
    
    # Initialize SVM & pipeline (Standardizing mean & variance)
    clf = make_pipeline(StandardScaler(),LinearSVC(random_state=0, tol=1e-4))

    # Train SVM
    clf.fit(X_train_new, Y_train)
    
    # Plot Confusion Matrix using trained classifier on Test Set.
    plot_confusion_matrix(clf, X_test_new, Y_test)
    

    # Predict on Testing set
    y_pred = clf.predict(X_test_new)

    # Printing the data-size (i.e. i)
    print('Printing Accuracy & ROC for '+ str(10 + 5*ind) +'% of data-size')
    
    # Calculate & print 'Accuracy' score using the predicted values
    print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))
    
    # Calculate & print 'ROC' score using the predicted values
    roc = multi_label_roc(Y_test, y_pred)
    print("ROC:",roc)
    
```

```python
# Approximate number of components (out of total 55) in order to get 10%,15%,20%,25%, & 30% of the originial data-size
n_comps = [5,7,10,13,15]

# Loop to iterate through different dimensionality reduction
for ind,i in enumerate(n_comps):
    
    # Initialize PCA with number of components ='5*i' (i.e. Approximately 10%,15%,20%,25%, & 30% of the originial 55 components)
    model = PCA(n_components=i)
    model.fit(X_train)
    X_train_new = model.transform(X_train)
    X_test_new = model.transform(X_test)
    
    # Initialize SVM & pipeline (Standardizing mean & variance)
    clf = make_pipeline(StandardScaler(),SVC(kernel="poly", degree=2, C=1))

    # Train SVM
    clf.fit(X_train_new, Y_train)
    
    # Plot Confusion Matrix using trained classifier on Test Set.
    plot_confusion_matrix(clf, X_test_new, Y_test)
    

    # Predict on Testing set
    y_pred = clf.predict(X_test_new)

    # Printing the data-size (i.e. i)
    print('Printing Accuracy & ROC for '+ str(10 + 5*ind) +'% of data-size')
    
    # Calculate & print 'Accuracy' score using the predicted values
    print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))
    
    # Calculate & print 'ROC' score using the predicted values
    roc = multi_label_roc(Y_test, y_pred)
    print("ROC:",roc)
    
```

**Model comparison**:

We logically knew that training the same model on a dimensionally reduced data-set can't provide more accurate results in comparison to training on original data.

For **Linear SVM**:

SVM (Original Data) gave: 'Accuracy': 0.81    'ROC (AUC)': 0.79

In comparision:

SVM (30% Data) gave: 'Accuracy': 0.79    'ROC (AUC)': 0.77

Hence our hypothesis of very few Principal components embedding most information is correct as the difference b/w SVM with (original data) & (30% data) is very low.

We can also compare lower models:

SVM (25% Data) also gives similar results to SVM (30% data). <br>
However, After that ROC start decreasing faster for lower dimension data. (i.e. SVM (5% data) gives ROC (AUC): 0.57)

For **Kernel SVM**:

Results are similar for kernel SVM. Infact of **Kernel SVM**, **SVM(30% data) performs as good as SVM(original data) with 0.84 of accuracy & ROC.**

Thus, we should try to find the sweet spot for the best % of Dimensionality reduction.

For finding that particular 'sweet spot', we can take the help of a **'SCREE Plot'**

```python
# Calculate variance in Principal components
scaler = StandardScaler()
pca = PCA()
pca_process = make_pipeline(scaler,pca)
pca_process.fit(X_train)

# Plotting Principal components & their variance (i.e. SCREE plot)
plt.bar(range(pca.n_components_), pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.show()
```

You can can observe that first **6-8 Principal components** carry most of the variance. These Principal components carry most of the embedded information in the data-set. This gives us an good idea about the ideal 'number of components' to choose during PCA.
